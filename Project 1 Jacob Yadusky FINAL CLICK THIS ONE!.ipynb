```python
import pandas as pd

#anything that is NaN is solved via errors = 'coerce'
df['Cores'] = pd.to_numeric(df['Cores'], errors='coerce')

high_end_threshold = 8
middle_end_threshold = 4

#we didnt go over 'pd.cut' but I learned it is used to bin data into differnt intervals such as middle, low and high.
df['Category'] = pd.cut(df['Cores'], bins=[0, middle_end_threshold, high_end_threshold, float('inf')],
                        labels=['Low End', 'Middle End', 'High End'], right=False)

df[['Name', 'Cores', 'Category']]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Category', order=['Low End', 'Middle End', 'High End'], palette='viridis')

plt.xlabel('Category')
plt.ylabel('Count')
plt.title('Distribution of CPUs by Category')

plt.show()
```

```python
#I googled this one line below where it says "r'(\d+\. \d+), I fell into a reddit thread and just started trying things until they worked."
df['Clock'] = pd.to_numeric(df['Clock'].str.extract(r'(\d+\.\d+)')[0], errors='coerce')

high_clock_threshold = 4.0
middle_clock_threshold = 3.0

df['Clock_Category'] = pd.cut(df['Clock'], bins=[0, middle_clock_threshold, high_clock_threshold, float('inf')],
                              labels=['Low Clock', 'Middle Clock', 'High Clock'], right=False)

df[['Name', 'Clock', 'Clock_Category']]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

#style
sns.set(style="whitegrid")

#figure and countplot with info inside
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Clock_Category', order=['Low Clock', 'Middle Clock', 'High Clock'], palette='viridis')

#labels
plt.xlabel('Clock Category')
plt.ylabel('Count')
plt.title('Distribution of CPUs by Clock Category')

#display it
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))

plt.scatter(df['Name'], df['Clock'], c='blue', label='Clock Speed', alpha=0.5)

#labels and title
plt.xlabel('CPU Names')
plt.ylabel('Clock Speed (GHz)')
plt.title('Clock Speeds of CPUs')
plt.xticks(rotation=90, ha='right')

#display
plt.tight_layout()
plt.show()
```
